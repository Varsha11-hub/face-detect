```python
import cv2
from matplotlib import pyplot as plt
from google.colab import files
```

```python
image_file='/content/OIP.webp'
image = cv2.imread(image_file)
```

```python
image=cv2.imread(image_file)
image_rgb=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
# convert for proper display

#cv2.imread()->reads the image as an array(openCV works with array)
#OpenCV reads BGR fromat by default.But matplotlib uses RGB format,so we convert it
```

```python
!wget -O haar_face.xml https://github.com/opencv/opencv/raw/master/data/haarcascades/haarcascade_frontalface_default.xml
face_cascade=cv2.CascadeClassifier('haar_face.xml')

#wget ->command-line tool to download files from the internet,
#-O haa-face.xml->save the downloaded files localy with the name haar_face.xml
#The URL ->pout to opencv's pre trained haar cascade file for frontal face detection
#Haar cascade-> pre-trained model to detect faces
#CascadeClassifier()-> load load the model so we can use it
```

```python
gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
faces=face_cascade.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=5)
```

```python
for (x,y,w,h) in  faces:
  cv2.rectangle(image_rgb,(x,y),(x+w,y+h),(255,0,0),2)

  #(x,y,w,h)->coordinates of detected face
  #cv2.rectangle()->draws a rectangle on the image
  #(255,0,0)->blue rectangle in RGB
  #2->thickness of the rectangle
```

```python
plt.figure(figsize=(8,6))
plt.imshow(image_rgb)
plt.title(f'Faces detected: {len(faces)}')
plt.axis('off')
plt.show()
```
